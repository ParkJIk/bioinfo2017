```python
gamakseq = open('Gamak_result.txt','r')
taxonomy = open('GreenGene_135.79_otus_taxonomy.txt','r')
humanflyworm = open('HumanFlyWorm_result.txt','r')
compare = {}
print('#Gamak Pond Sequencing Results', '\n')

for tax in taxonomy:
    confirm = tax.split()
    compare[confirm[0]] = tax
    
for line1 in gamakseq:
    if line1.startswith('#'):
        stack1 = 1
        continue
    else:
        if stack1 == 1:
            FirstLine = line1.split()
            query1 = FirstLine[0]
            subjt1 = FirstLine[1]
            ident1 = FirstLine[2]
            stack1 = 0
            print(' Sequence: ', query1, '\n', 'Accuracy: ', ident1, '%')
            
            species = compare[subjt1].split()
            number = species[0]
            k_species = species[1]
            p_species = species[2]
            c_species = species[3]
            o_species = species[4]
            f_species = species[5]
            g_species = species[6]
            s_species = species[7]
            
            print(' Classify: ', number)
            print('             ', k_species)
            print('             ', p_species)
            print('             ', c_species)
            print('             ', o_species)
            print('             ', f_species)
            print('             ', g_species)
            print('             ', s_species, '\n')
            

print('\n', '\n', '#Human and Fly ensulin compare', '\n')
for line2 in humanflyworm:
    if line2.startswith('#'):
        stack2 = 1
        continue
    else:
        if stack2 == 1:
            SecondLine = line2.split()
            query2 = SecondLine[0]
            subjt2 = SecondLine[1]
            ident2 = SecondLine[2]
            alilen = SecondLine[3]
            mismch = SecondLine[4]
            gapopn = SecondLine[5]
            qstart = SecondLine[6]
            qend = SecondLine[7]
            sstart = SecondLine[8]
            send = SecondLine[9]
            evalue = SecondLine[10]
            bscore = SecondLine[11]
            stack2 = 0
            print('Query Acc.ver:        ', query2)
            print('Subject Acc.ver:      ', subjt2)
            print('% Identity:           ', ident2)
            print('Alignment Length:     ', alilen)
            print('Number of Mismatches: ', mismch)
            print('Number of Gap-opens:  ', gapopn)
            print('Query Start:          ', qstart)
            print('Query End:            ', qend)
            print('Subject Start:        ', sstart)
            print('Subject End:          ', send)
            print('E-value:              ', evalue)
            print('Bit Score:            ', bscore, '\n')
```
